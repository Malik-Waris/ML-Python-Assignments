```python

from keras.applications.vgg16 import VGG16

model = VGG16(weights='imagenet')
print(model.summary())
```

All pre-trained models expect input images normalized in the same way,
i.e. mini-batches of 3-channel RGB images of shape `(3 x H x W)`, where `H` and `W` are expected to be at least `224`.
The images have to be loaded in to a range of `[0, 1]` and then normalized using `mean = [0.485, 0.456, 0.406]`
and `std = [0.229, 0.224, 0.225]`.

Here's a sample execution.

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions
import numpy as np
value=10
arr = [0] * value
path=[0]*value
img=[0]*value
for n in range(value):
    arr[n] = '/content/sample_data/do'+str(n+1)+'.PNG'
for n in range(value):
    path[n] =arr[n] 
    img[n] = image.load_img(path[n],color_mode='rgb', target_size=(224, 224))
    display(img[n])
#There is an interpolation method to match the source size with the target size
#image loaded in PIL (Python Imaging Library)
```

```python
# Converts a PIL Image to 3D Numy Array

#0
x = image.img_to_array(img[0])
x.shape
# Adding the fouth dimension, for number of images
x = np.expand_dims( x, axis=0)
#1
x1 = image.img_to_array(img[1])
x1.shape
# Adding the fouth dimension, for number of images
x1 = np.expand_dims( x1, axis=0)
#2
x2 = image.img_to_array(img[1])
x2.shape
# Adding the fouth dimension, for number of images
x2 = np.expand_dims( x2, axis=0)
#3
x3 = image.img_to_array(img[3])
x3.shape
# Adding the fouth dimension, for number of images
x3 = np.expand_dims(x3 ,axis=0)
#4
x4 = image.img_to_array(img[4])
x4.shape
# Adding the fouth dimension, for number of images
x4 = np.expand_dims( x4, axis=0)
#5
x5 = image.img_to_array(img[5])
x5.shape
# Adding the fouth dimension, for number of images
x5 = np.expand_dims( x5, axis=0)
#6
x6 = image.img_to_array(img[6])
x6.shape
# Adding the fouth dimension, for number of images
x6 = np.expand_dims( x6, axis=0)
#7
x7 = image.img_to_array(img[7])
x7.shape
# Adding the fouth dimension, for number of images
x7 = np.expand_dims( x7, axis=0)
#8
x8 = image.img_to_array(img[8])
x8.shape
# Adding the fouth dimension, for number of images
x8 = np.expand_dims( x8, axis=0)
#9
x9 = image.img_to_array(img[9])
x9.shape
# Adding the fouth dimension, for number of images
x9 = np.expand_dims( x9, axis=0)
```

```python
#mean centering with respect to Image

x = preprocess_input(x)
features = model.predict(x)
p = print(decode_predictions(features))
x1 = preprocess_input(x1)
features = model.predict(x1)
p = print(decode_predictions(features))
x2 = preprocess_input(x2)
features = model.predict(x2)
p = print(decode_predictions(features))
x3 = preprocess_input(x3)
features = model.predict(x3)
p = print(decode_predictions(features))
x4 = preprocess_input(x4)
features = model.predict(x4)
p = print(decode_predictions(features))
x5 = preprocess_input(x5)
features = model.predict(x5)
p = print(decode_predictions(features))
x6 = preprocess_input(x6)
features = model.predict(x6)
p = print(decode_predictions(features))
x7 = preprocess_input(x7)
features = model.predict(x7)
p = print(decode_predictions(features))
x8 = preprocess_input(x8)
features = model.predict(x8)
p = print(decode_predictions(features))
x9 = preprocess_input(x9)
features = model.predict(x9)
p = print(decode_predictions(features))
```
